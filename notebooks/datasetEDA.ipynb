```python
import os
import pandas as pd
import orjson
from tqdm import tqdm

data_path = "../turkey-gas-station-data/raw-json"

all_stations = []
json_files = [f for f in os.listdir(data_path) if f.endswith(".json")]

for file in tqdm(json_files):
    provider_name = file.split('.')[0].capitalize()
    file_path = os.path.join(data_path, file)
    
    with open(file_path, 'rb') as f:
        data = orjson.loads(f.read())
        
        if isinstance(data, dict):
            data = [data]
            
        for item in data:
            item['source_provider'] = provider_name
            all_stations.append(item)

df = pd.DataFrame(all_stations)
```

```python
po_raw = df[df['source_provider'] == 'Petrol_ofisi'].copy()
others = df[df['source_provider'] != 'Petrol_ofisi'].copy()

po_exploded = po_raw.dropna(subset=['Values']).explode('Values')
po_flattened = pd.json_normalize(po_exploded['Values'])
po_flattened['source_provider'] = 'Petrol_ofisi' # Marka adını geri eklemeyi unutma

# PO ile diğer markaları tek bir tabloda (ndf) alt alta birleştiriyoruz
ndf = pd.concat([others, po_flattened], ignore_index=True)
```

```python
import re

def clean_shell_district(row):
    if pd.isna(row['city']) or pd.isna(row['state']):
        return row['city']
        
    dist_str = str(row['city']).upper()
    city_str = str(row['state']).upper()
    
    # İl adının hem orijinalini hem de olası Türkçe karakterli halini listele
    # (Örn: MUGLA ve MUĞLA, IZMIR ve İZMİR)
    city_str_tr = city_str.replace('G', 'Ğ').replace('I', 'İ').replace('C', 'Ç').replace('S', 'Ş')
    
    # İki varyasyonu da ilçe metninden sil
    dist_str = dist_str.replace(city_str, '').replace(city_str_tr, '')
    
    # Sondaki veya baştaki gereksiz boşlukları ve tire (-) işaretlerini Regex ile temizle
    dist_str = re.sub(r'^[-\s]+|[-\s]+$', '', dist_str).strip()
    
    return dist_str

# Shell maskesi ve uygulanması
shell_mask = ndf['source_provider'] == 'Shell'
if shell_mask.any():
    ndf.loc[shell_mask, 'unified_city'] = ndf.loc[shell_mask, 'state']
    # Artık akıllı fonksiyonumuzu kullanıyoruz
    ndf.loc[shell_mask, 'unified_district'] = ndf.loc[shell_mask].apply(clean_shell_district, axis=1)
```

```python
lat_candidates = ["Enlem", "Lat", "lat", "latitude", "Latitude"]
lon_candidates = ["Boylam", "Lon", "lng", "longitude", "lon", "Lng", "Longitude"]

valid_lat_cols = [col for col in lat_candidates if col in ndf.columns]
valid_lon_cols = [col for col in lon_candidates if col in ndf.columns]

ndf['unified_lat'] = ndf[valid_lat_cols].bfill(axis=1).iloc[:, 0]
ndf['unified_lon'] = ndf[valid_lon_cols].bfill(axis=1).iloc[:, 0]
```

```python
def standardized_texts(text):
    if pd.isna(text):
        return text
        
    text = str(text).strip().upper()
    
    changes = {
        'DIYARBAKIR': 'DİYARBAKIR',
        'BINGOL': 'BİNGÖL',
        'BITLIS': 'BİTLİS',
        'DENIZLI': 'DENİZLİ',
        'DUZCE': 'DÜZCE',
        'ISTANBUL': 'İSTANBUL',
        'IZMIR': 'İZMİR',
        'AFYONKARAHISAR': 'AFYONKARAHİSAR',
        'ARTVIN': 'ARTVİN',
        'AGRI': 'AĞRI',
        'BALIKESIR': 'BALIKESİR',
        'BINGOL': 'BİNGÖL',
        'BILECIK': 'BİLECİK',
        'BITLIS': 'BİTLİS',
        'EDIRNE': 'EDİRNE',
        'ELAZIG': 'ELAZIĞ',
        'ERZINCAN': 'ERZİNCAN',
        'ESKISEHIR': 'ESKİŞEHİR',
        'GAZIANTEP': 'GAZİANTEP',
        'GIRESUN': 'GİRESUN',
        'GUMUSHANE': 'GÜMÜŞHANE',
        'HAKKARI': 'HAKKARİ',
        'IGDIR': 'IĞDIR',
        'KARABUK': 'KARABÜK',
        'BITLIS': 'BİTLİS',
        'KAYSERI': 'ELAZIĞ',
        'KIRKLARELI': 'ERZİNCAN',
        'KIRSEHIR': 'ESKİŞEHİR',
        'KOCAELI': 'GAZİANTEP',
        'KUTAHYA': 'GİRESUN',
        'MARDIN': 'GÜMÜŞHANE',
        'MERSIN': 'HAKKARİ',
        'MUGLA': 'IĞDIR',
        'MUS': 'KARABÜK',
        'MANISA': 'BİTLİS',
        'MERSIN': 'ESKİŞEHİR',
        'NEVSEHIR': 'GAZİANTEP',
        'NIGDE': 'GİRESUN',
        'OSMANIYE': 'GÜMÜŞHANE',
        'RIZE': 'HAKKARİ',
        'SANLIURFA': 'IĞDIR',
        'SIIRT': 'KARABÜK',
        'SINOP': 'BİTLİS',
        'SIRNAK': 'IĞDIR',
        'SIVAS': 'KARABÜK',
        'MANISA': 'BİTLİS',
        'TEKIRDAG': 'ESKİŞEHİR',
        'TUNCELI': 'GAZİANTEP',
        'USAK': 'GİRESUN',
        'OSMANIYE': 'GÜMÜŞHANE',
        'RIZE': 'HAKKARİ',
        'SANLIURFA': 'IĞDIR',
        'SIIRT': 'KARABÜK',
        'SINOP': 'BİTLİS',
    }
    
    for old, new in changes.items():
        text = text.replace(old, new)
        
    # Title Case format
    # Python'un standart .title() fonksiyonu İ/I harflerini bozduğu için özel bir formatlama yapıyoruz:
    if len(text) > 1:
        return text[0] + text[1:].replace('I', 'ı').replace('İ', 'i').lower()
    return text


silver_df['unified_city'] = silver_df['unified_city'].apply(standardized_texts)
silver_df['unified_district'] = silver_df['unified_district'].apply(standardized_texts)
```
